````python
def listReg(x):
    regex = r"\<(.*?)\>"
    matches = re.finditer(regex, x,re.MULTILINE)
    tagList = []
    for matchNum, match in enumerate(matches, start = 1):
        #print ("Match {matchNum} was found at {start}-{end}: {match}".format(matchNum = matchNum, start = match.start(), end = match.end(), match = match.group()))
        #print("matchNum", matchNum)
        #print("match", match)
        for groupNum in range(0, len(match.groups())):
            #print(match.group(groupNum))
            groupNum = groupNum + 1
            tagList.append(match.group(groupNum))
            #print ("Group {groupNum} found at {start}-{end}: {group}".format(groupNum = groupNum, start = match.start(groupNum), end = match.end(groupNum), group = match.group(groupNum)))
    return tagList if match else None

def inputTransf(x, result = "Token"):
    html = x["Body"]
    raw = BeautifulSoup(html, "html.parser").get_text()
    text = "[CLS] " + raw + " [SEP]"
    tokenized_text_1 = tokenizer.tokenize(text)
    tokenized_text_1 = tokenized_text_1[:512]
    indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text_1)
    segments_ids = [0] * len(indexed_tokens)
    #print(segments_ids)

    if(type(x["Tags"]) != np.float):
        text2 = listReg(x["Tags"])
        len_text2 = len(text2)
        text2 = ' '.join(text2)

        final_text = text + text2 + " [SEP]"
        final_token = tokenizer.tokenize(final_text)
        final_token = final_token[:512]
        indexed_tokens_2 = tokenizer.convert_tokens_to_ids(final_token)
        segments_mark = [1] * len(indexed_tokens_2)

        final_segments = []
        final_indexes = []

        final_segments.append(segments_ids[:])
        final_segments.append(segments_mark[:])
        final_segments = [item for sub in final_segments for item in sub]

        final_indexes.append(indexed_tokens[:])
        final_indexes.append(indexed_tokens_2[:])
        final_indexed_token = [item for sub in final_indexes for item in sub]

        if(len(final_segments) != len(final_indexed_token)): print("chou")
        if(result == "Token"): return torch.tensor(final_indexed_token).unsqueeze(1).to("cuda")
        else : return torch.tensor(final_segments).unsqueeze(1).to("cuda")

        #return final_indexed_token, final_segments
    if(len(indexed_tokens) != len(segments_ids)): print("ouch")
    if (result == "Token") : return torch.tensor(indexed_tokens).unsqueeze(1).to("cuda")
    else : return torch.tensor(segments_ids).unsqueeze(1).to("cuda")
    #return indexed_tokens, segments_ids

def input(x, result = "Token"):
    html = x["Body"]
    raw = BeautifulSoup(html, "html.parser").get_text()
    text = "[CLS] " + raw + " [SEP]"
    tokenized_text_1 = tokenizer.tokenize(text)
    tokenized_text_1 = tokenized_text_1[:512]
    indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text_1)
    segments_ids = [0] * len(indexed_tokens)
    #print(segments_ids)

    if(result == "Token"): return indexed_tokens
    else : return segments_ids
    #return indexed_tokens, segments_ids
````

````python
 token_embeddings = torch.stack(encoded_layers, dim=0)

token_embeddings.size()
````

````python
# Remove dimension 1, the "batches".
token_embeddings = token_embeddings.permute(2,0,1,3)
print(token_embeddings.size())
token_embeddings = torch.squeeze(token_embeddings)
print(token_embeddings.size())
# Swap dimensions 0 and 1.
token_embeddings = token_embeddings.permute(1,0,2)

print(token_embeddings.size())
#torch.Size([1, 12, 328, 768])
#torch.Size([12, 328, 768])
#torch.Size([328, 12, 768])
````

````python
# Stores the token vectors, with shape [22 x 3,072]
token_vecs_cat = []

# `token_embeddings` is a [22 x 12 x 768] tensor.

# For each token in the sentence...
for token in token_embeddings:
    
    # `token` is a [12 x 768] tensor

    # Concatenate the vectors (that is, append them together) from the last 
    # four layers.
    # Each layer vector is 768 values, so `cat_vec` is length 3,072.
    cat_vec = torch.cat((token[-1], token[-2], token[-3], token[-4]), dim=0)
    
    # Use `cat_vec` to represent `token`.
    token_vecs_cat.append(cat_vec)

print ('Shape is: %d x %d' % (len(token_vecs_cat), len(token_vecs_cat[0])))
#Shape is: 14 x 3072
````

````python
# `encoded_layers` has shape [12 x 1 x 22 x 768]

# `token_vecs` is a tensor with shape [22 x 768]
token_vecs = encoded_layers[10][0]

# Calculate the average of all 22 token vectors.
sentence_embedding = torch.mean(token_vecs, dim=0)

print ("Our final sentence embedding vector of shape:", sentence_embedding.size())
````

````python
import tensorflow_hub as hub
import tensorflow as tf

elmo = hub.Module("https://tfhub.dev/google/elmo/2", trainable=True)
````

```
# just a random sentence
x = ["Roasted ants are a popular snack in Columbia"]

# Extract ELMo features 
embeddings = elmo(x, signature="default", as_dict=True)["elmo"]

embeddings.shape
```
The output is a 3 dimensional tensor of shape (1, 8, 1024):

The first dimension of this tensor represents the number of training samples. This is 1 in our case
The second dimension represents the maximum length of the longest string in the input list of strings. Since we have only 1 string in our input list, the size of the 2nd dimension is equal to the length of the string – 8
The third dimension is equal to the length of the ELMo vector

````python
def elmo_vectors(x):
  embeddings = elmo(x, signature="default", as_dict=True)["elmo"]

  with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    sess.run(tf.tables_initializer())
    # return average of ELMo features
    return sess.run(tf.reduce_mean(embeddings,1))
````
